# Create continuous income level

# Include information in teachers dataset

```python
platforms = ['italki', 'preply', 'verbling']
country_codes = pd.read_csv('../data/income_level_code.csv', index_col=0).drop(columns=['Country Name']).rename(columns={'gdp_capita': 'income_level'})

for platform in platforms:
    path = '../data/results/final_dataframes/{}.csv'.format(platform)
    df = pd.read_csv(path, index_col=0)
    df = df.drop(columns=['income_level', 'Code'])
    df = df.merge(country_codes, left_on='nationality', right_on='Code', how='left')
    if len(df[df['income_level'].isna()]):
        print(df[df['income_level'].isna()].nationality.unique())
    df.to_csv('../data/results/final_dataframes/{}.csv'.format(platform))
```
